```python
import pickle
import tweepy 
from twitter_keys import consumer_key ,consumer_secret,access_token, access_secret
#import pickle
auth =tweepy.OAuthHandler(consumer_key,consumer_secret)
auth.set_access_token(access_token,access_secret)
api=tweepy.API(auth)
public_tweets1 =api.user_timeline('RGVzoomin')
with open ('rgv.pkl','wb') as f:
    pickle.dump(public_tweets1,f)
public_tweets2 =api.user_timeline('imVkohli')
with open ('virat.pkl','wb') as f:
    pickle.dump(public_tweets2,f) 
public_tweets3 =api.user_timeline('zaynmalik')
with open ('zayn.pkl','wb') as f:
    pickle.dump(public_tweets3,f) 
```

```python
with open('rgv.pkl','rb')as f:
    my_data1 =pickle.load(f) 
    print(my_data1)
with open('virat.pkl','rb')as f:
    my_data2 =pickle.load(f)
    print(my_data2)
with open('zayn.pkl','rb')as f:
    my_data3 =pickle.load(f)  
    print(my_data3)
```

## Description of Analysis Steps 

###  1. Lexical Diversity.
###  2. commonly used words in tweets.
###  3.  Devices Used for the twitter.
###  4. Punctuation/capitalization usage in tweets .
###  5. Sentimental Analysis using matplotlib.
###  6. Average Diversities for the tweets.

```python
#lexical diversity

def lexical_diversity(text):
    tokens= text.lower().split()
    return len(set(tokens))/len(tokens)

adiversities =[]
for tweet in my_data1:
    adiversities.append(lexical_diversity(tweet.text))
print(adiversities)

bdiversities =[]
for tweet in my_data2:
    bdiversities.append(lexical_diversity(tweet.text))
print(bdiversities)

cdiversities =[]
for tweet in my_data3:
    cdiversities.append(lexical_diversity(tweet.text))
print(cdiversities)


```

```python
#Commonly used words

from collections import Counter
public_tweets1= api.user_timeline('RGVzoomin')
for tweet in public_tweets1:
    words1=tweet.text.lower().split()
    c=Counter(words1)
    print(c)
    print('The commonly used word')
    print(c.most_common(1))
    print('-------------------------')
public_tweets2= api.user_timeline('imVkohli')
for tweet in public_tweets1:
    words2=tweet.text.lower().split()
    c=Counter(words2)
    print(c)
    print('The commonly used word')
    print(c.most_common(1)) 
    print('-------------------------')
public_tweets3= api.user_timeline('zaynmalik')
for tweet in public_tweets1:
    words3=tweet.text.lower().split()
    c=Counter(words3)
    print(c)
    print('The commonly used word')
    print(c.most_common(1)) 
    print('-------------------------')
```

```python
#Devices used

#rgv devices
rgv_devices = []
for tweet in public_tweets1:
    rgv_devices.append(tweet.source)
print('Devices for rgv: ',rgv_devices)
print()

#kohli devices

kohli_devices = []
for tweet in public_tweets2:
    kohli_devices.append(tweet.source)
print('Devices for kohli: ',kohli_devices)
print()

# zayn devices

zayn_devices = []
for tweet in public_tweets3:
    zayn_devices.append(tweet.source)
print('Devices for zayn : ',zayn_devices)
```

```python
# Tweet Length
for t in public_tweets1:
    tokens =t.text.split()
    print(tokens)
    print(len(tokens))
    print('------------------')
for t in public_tweets2:
    tokens =t.text.split()
    print(tokens)
    print(len(tokens))
    print('------------------')  
for t in public_tweets3:
    tokens =t.text.split()
    print(tokens)
    print(len(tokens))
    print('------------------')    
```

```python
#Punctuation/capitalization usage

import nltk
nltk.download('vader_lexicon')

import numpy as np
from nltk.sentiment.vader import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
for tweet in public_tweets1:
    print('Rgv tweet analyzer')
    print(analyzer.polarity_scores(tweet.text))
    print('--------')
for tweet in public_tweets2:
    print('virat kohli tweet analyzer')
    print(analyzer.polarity_scores(tweet.text))
    print('---------')
for tweet in public_tweets3:
    print('zayn tweet analyzer')
    print(analyzer.polarity_scores(tweet.text))
    print('-----------')
    
    
    
    



```

```python
#Sentimental Analysis

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from nltk_helper import get_sentiments


fig = plt.figure()

ax=fig.add_subplot(111,projection='3d')

sentiments = [get_sentiments(tweet.text) for tweet in public_tweets1]


xs=[]
for sent in sentiments:
    xs.append(sent['neg'])
    
ys=[sent['neu'] for sent in sentiments]

zs = list(map(lambda x:x['pos'],sentiments))

ax.scatter(xs,ys,zs)



    
from nltk_helper import split_sentiments

vX,vY,vZ = split_sentiments([get_sentiments(tweet.text) for tweet in public_tweets2])
ax.scatter(vX,vY,vZ, color='r',marker='*')
#ax.scatter(arbyX,arbyY,arbyZ)

#ax.scatter(xs,ys,zs)
ax.set_xlabel('Negative')
ax.set_ylabel('neutral')
ax.set_zlabel('postive')




    
from nltk_helper import split_sentiments

zX,zY,zZ = split_sentiments([get_sentiments(tweet.text) for tweet in public_tweets3])
ax.scatter(zX,zY,zZ, color='g',marker='d')
#ax.scatter(arbyX,arbyY,arbyZ)

#ax.scatter(xs,ys,zs)
ax.set_xlabel('Negative')
ax.set_ylabel('neutral')
ax.set_zlabel('postive')

plt.show()

```

```python
# Average diversities

print('RGV average diversities:',sum(adiversities)/len(adiversities))
print('Virat average diversities :',sum(bdiversities)/len(bdiversities))
print('zayn average diversities :',sum(cdiversities)/len(cdiversities))
```
